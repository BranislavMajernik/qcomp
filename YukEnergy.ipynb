```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for N in range(2,5):
    x  = np.arange(0, 40, 0.1)
    y1 = (np.sin(x/2*N))/(np.sin(x/2))
    y= np.exp(np.array([1j])*x*(N-1)/2)*(np.sin(x/2*N))/(np.sin(x/2))
    plt.plot(x,y, color='blue')
    plt.grid(True, which='both')
    plt.title(N)
    #peaks = np.where((y[1:-1] > y[0:-2]) * (y[1:-1] > y[2:]))[0] + 1
    dips = np.where((y[1:-1] < y[0:-2]) * (y[1:-1] < y[2:]))[0] + 1
    plt.plot (x, y)
    print(x[dips], np.exp(x[dips]))
    #plt.plot (x[peaks], y[peaks], 'o')
    plt.plot (x[dips], y[dips], 'o')
    plt.show()
```
